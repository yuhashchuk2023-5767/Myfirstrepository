Завдання задано у попередній комірці.

Всі коментарі та вивід у функції `print` мають бути українською мовою.  
Завжди створюй новий код програми.


```python
class UniquenessChecker:
    def __init__(self, elements: list):
        # Зберігаємо список елементів
        self.elements = elements
        self._is_unique = None

    def check(self):
        # Перевіряємо, чи всі елементи унікальні
        self._is_unique = len(self.elements) == len(set(self.elements))
        return self._is_unique

    @property
    def is_unique(self):
        # Повертаємо результат перевірки
        if self._is_unique is None:
            self.check()
        return self._is_unique

    def __str__(self):
        # Вивід інформації про студента
        return "Ім'я студента: Богдан, Група: КН-31"


# Створення декількох об'єктів класу з різними вхідними даними
checkers = [
    UniquenessChecker([1, 2, 3, 4]),
    UniquenessChecker([1, 2, 2, 3]),
    UniquenessChecker(["apple", "banana", "cherry"])
]

for checker in checkers:
    print(checker)
    print(f"Список елементів: {checker.elements}")
    print(f"Чи всі елементи унікальні: {checker.is_unique}")
    print("-" * 40)
```
