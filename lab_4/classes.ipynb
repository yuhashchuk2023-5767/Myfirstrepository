```python
# Cell 1 : Enum та імпорти
from enum import Enum
import random

class AnimalStatus(Enum):
    ACTIVE = "активна"
    SLEEPING = "спить"
    EATING = "їсть"
```

```python
#Cell 2 : Класи
class Animal:
    def __init__(self, species, name, age):
        self.species = species    # Вид тварини
        self.name = name          # Ім'я
        self.age = age            # Вік
        self.status = AnimalStatus.ACTIVE

class Owner:
    def __init__(self, name):
        self.name = name
        self.animals = []         # Список тваринок

class Statistics:
    def __init__(self):
        self.activity_log = {}
        self.popularity = {}
```

```python
# Cell 3 : Клас PetCenter
class PetCenter:
    def __init__(self):
        self.animals = []
        self.owners = []
        self.stats = Statistics()

    def add_animal(self, animal):
        self.animals.append(animal)

    def add_owner(self, owner):
        self.owners.append(owner)

    def give_animal(self, owner, animal):
        owner.animals.append(animal)
        
        # Статистика
        self.stats.activity_log[owner.name] = \
            self.stats.activity_log.get(owner.name, 0) + 1

        self.stats.popularity[animal.species] = \
            self.stats.popularity.get(animal.species, 0) + 1

    def change_status(self, animal):
        animal.status = random.choice(list(AnimalStatus))

    def print_statistics(self):
        print("Активність власників:")
        for owner, count in self.stats.activity_log.items():
            print(f"{owner}: {count} тваринок")

        print("\nПопулярність видів:")
        for species, count in self.stats.popularity.items():
            print(f"{species}: {count} разів")
```

```python
#Cell 4 : Симуляція
def simulate_days(center, days=60):
    for _ in range(days):
        owner = random.choice(center.owners)
        animal = random.choice(center.animals)
        center.give_animal(owner, animal)
        center.change_status(animal)
```

```python
#Cell 5 : Виконання 
center = PetCenter()

# Додаємо тварин
center.add_animal(Animal("Собака", "Берта", 4))
center.add_animal(Animal("Хом'як", "Джексон", 1))
center.add_animal(Animal("Кіт", "Кіті", 3))

# Додаємо власників
center.add_owner(Owner("Юля"))
center.add_owner(Owner("Андрій"))
center.add_owner(Owner("Марко"))

# Запускаємо симуляцію
simulate_days(center)

# Друкуємо статистику
center.print_statistics()
```
