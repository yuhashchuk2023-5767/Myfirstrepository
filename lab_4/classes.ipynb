```python
# Cell 1 : Enum та імпорти
from enum import Enum
import random

class AnimalStatus(Enum):
    ACTIVE = "активна"
    SLEEPING = "спить"
    EATING = "їсть"
```

```python
#Cell 2 : Класи
class Animal:
    def __init__(self, species, name, age):
        self.species = species    # Вид тварини
        self.name = name          # Ім'я
        self.age = age            # Вік
        self.status = AnimalStatus.ACTIVE

class Owner:
    def __init__(self, name):
        self.name = name
        self.animals = []         # Список тваринок

class Statistics:
    def __init__(self):
        self.activity_log = {}
        self.popularity = {}
```

```python
# Cell 3 : Клас PetCenter
class PetCenter:
    def __init__(self):
        self.animals = []
        self.owners = []
        self.stats = Statistics()

    def add_animal(self, animal):
        self.animals.append(animal)

    def add_owner(self, owner):
        self.owners.append(owner)

    def give_animal(self, owner, animal):
        owner.animals.append(animal)
        
        # Статистика
        self.stats.activity_log[owner.name] = \
            self.stats.activity_log.get(owner.name, 0) + 1

        self.stats.popularity[animal.species] = \
            self.stats.popularity.get(animal.species, 0) + 1

    def change_status(self, animal):
        animal.status = random.choice(list(AnimalStatus))

    def print_statistics(self):
        print("Активність власників:")
        for owner, count in self.stats.activity_log.items():
            print(f"{owner}: {count} тваринок")

        print("\nПопулярність видів:")
        for species, count in self.stats.popularity.items():
            print(f"{species}: {count} разів")
```

```python
#Cell 4 : Симуляція
def simulate_days(center, days=60):
    for _ in range(days):
        owner = random.choice(center.owners)
        animal = random.choice(center.animals)
        center.give_animal(owner, animal)
        center.change_status(animal)
```

```python
#Cell 5 : Виконання 
center = PetCenter()

# Додаємо тварин
center.add_animal(Animal("Собака", "Берта", 4))
center.add_animal(Animal("Хом'як", "Вудді", 1))
center.add_animal(Animal("Кіт", "Кіті", 3))

# Додаємо власників
center.add_owner(Owner("Юля"))
center.add_owner(Owner("Андрій"))
center.add_owner(Owner("Марко"))

# Запускаємо симуляцію
simulate_days(center)

# Друкуємо статистику
center.print_statistics()
```

```python
#Cell 11 : Статичні методи та змінна self
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила зі мною один рік"

    @staticmethod
    def call_pet():
        return "Тваринка біжить до нас"


a1 = MyAnimals("Собака", "Берта", 3)
a2 = MyAnimals("Кіт", "Чарлі", 6)

print(f"Ми кличемо нашу тваринку за допомогою статичного методу з обєкту |{a1.call_pet()}| та з класу |{MyAnimals.call_pet()}|")

print(f"Викликаю метод з обєкту |{a1.add_one_year()}| та з класу |{MyAnimals.add_one_year(a1)}|")
```

```python
#Cell 12 : Робота з properties
# додайте ще одну properties та виведіть її
from datetime import datetime

class MyAnimals:
    max_age = 20

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    @property
    def live_untill(self):
        return datetime.today().year - self.age + self.max_age

    def add_one_year(self):
        self.age += 1
        return f"Моя Тваринка {self.animals} прожила зі мною один рік"


a1 = MyAnimals("Кіт", "Чарлі", 6)
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
a1.add_one_year()
print(f"Мій {a1.animals} буде зі мною до {a1.live_untill} року")
```

```python
#Cell 13 : Методи Класу
# допишіть метод класу який буде превіряти що атрибут name заданий з великої букви
class MyAnimals:
    total_csv = int(0)
    total_objects = int(0)

    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age
        MyAnimals.total_objects += 1

    @classmethod
    def from_csv(cls, csv_data: str):
        cls.total_csv += 1
        animal, name, age = csv_data.split(",")
        return cls(animal, name, age)


a1 = MyAnimals.from_csv("Собака,Берта,3")
a2 = MyAnimals("Кіт", "Чарлі", 6)

print(f"Моя {a1.animals}, звати {a1.name}, має {a1.age} роки")
print(f"Мій {a2.animals}, звати {a2.name}, має {a2.age} років")
print(f"Загальна кількість обєктів {MyAnimals.total_objects} а створених з csv формату {MyAnimals.total_csv}")
```

```python
#Cell 14 :Магічні методи
class MyAnimals:
    def __init__(self, animals, name, age) -> None:
        self.animals = animals
        self.name = name
        self.age = age

    # Офіційне представлення об'єкта
    def __repr__(self) -> str:
        return f'MyAnimals("{self.animals}", "{self.name}", {self.age})'

    # Зручне для користувача представлення
    def __str__(self) -> str:
        return f"Це є обєкт {self.__repr__()}"


a1 = MyAnimals("Собака", "Берта", 3)
print(a1)

print(repr(a1))

str(a1)
```
